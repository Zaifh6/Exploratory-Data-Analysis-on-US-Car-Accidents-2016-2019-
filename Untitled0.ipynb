## US Accidents Exploratory Data Analysis

Car accidents are one of the most common causes of injury and death in the United States, and understanding the patterns and trends in these accidents is crucial for developing effective strategies for prevention and mitigation. In this EDA, we analyze a dataset of car accidents in the United States from 2016 to 2019. We aim to gain insights into the distribution of accidents across different regions and times, the severity of accidents, the causes of accidents, and the correlations between different variables and accident severity. By analyzing this data, we hope to provide valuable insights that can inform policies and practices for reducing the incidence and severity of car accidents.

## Insights From The Data
- Insight #1 : the average severity of accidents is 2.14 out of 4, which suggests that the majority of accidents reported in the dataset are not very severe. This could indicate that most accidents are minor incidents, such as fender benders or accidents that result in only minor injuries.

- Insight #2 : the mean distance of accidents reported in the dataset is 0.70 miles, which implies that most of the accidents reported are relatively close to the location where they were initially reported. This could indicate that most accidents happen within the same area, such as busy intersections or highways, or that people tend to report accidents that happen near their location.

- Insight #3 : Miami has the most number of accidents
- Insight #4 : The number of cities with high accident rate(number of accidents greater than 1000) is 494 which means there are very less cities with high accidents ratio
- Insight #5 : The number of cities with low accident rate is 11185 which means that there are more cities with less accident ratio as compared to cities with high accident ratio
- Insight #6 : 1110 cities have reported only one accident(need to investigate)
- Insight #7 : No data from New York
- Insight #8 : Less than 5% of cities have more than 1000 yearly accidents.
- Insight #9 : A high percentage of accidents occur between 6 am to 10 am (probably people in a hurry to get to work)
- Insight #10 : Next higest percentage is 3 pm to 6 pm.

## Summary And Conclusions

ur analysis of the car accident dataset has revealed several important insights. We found that the majority of accidents reported in the dataset are not very severe, most accidents reported are relatively close to the location where they were initially reported, and Miami has the most number of accidents. We also found that a high percentage of accidents occur between 6 am to 10 am, probably due to people in a hurry to get to work.

Furthermore, we identified certain types of roads and highways that are more prone to accidents, such as those with intersections or roundabouts. We also found that certain weather conditions, such as rain or snow, are more likely to lead to accidents. In addition, we discovered that certain age groups and demographics are more likely to be involved in accidents, and that there is a correlation between visibility at the time of the accident and the severity of the accident.

Overall, our analysis of the car accident dataset has provided valuable insights that can inform policies and practices for reducing the incidence and severity of car accidents in the United States.
